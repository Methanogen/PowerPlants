```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv("gen1016m.csv",skiprows= 2, nrows=7, header= None, dtype = datatypes, usecols= range(84))
df1 = df.T
df1.head()
df1.rename(columns={0:'month', 1: 'ng',2:'coal',3:'nuclear',4:'wind',5:'sgen', 6:'gasprice'}, inplace=True)
df1.head()
df1['month'] = df1['month'].astype('category')
df1['ng'] = df1['ng'].astype('float')
df1['coal'] = df1['coal'].astype('float')
df1['nuclear'] = df1['nuclear'].astype('float')
df1['wind'] = df1['wind'].astype('float')
df1['sgen'] = df1['sgen'].astype('float')
df1['gasprice'] = df1['gasprice'].astype('float')

df1.head(3)

df1.info()
df1['month'].nunique()# now correctly returns 12 !!!
# df1['month'].head(50)
#df1.dtype
#df1.shape()
df1.ng = df1.ng/1000
df1.coal = (df1.coal/1000).astype(int)
df1.nuclear = (df1.nuclear/1000).astype(int)
df1.wind = (df1.wind/1000).astype(int)
df1.sgen = (df1.sgen/1000).astype(int)
df1.head(10)
#DWW later add time and month (as temp) in important "str as datatype, convert to sategory (One-hot encoding?)
```
