```python
# We seek to make a statement of the following form. In the state of Texas from 2010 to 2016 the power generated from 
# Wind Energy increased [ten]fold to x GWh. However, over this same period the Heat Rates for the type of gas-fired 
# power plants believed to back up the intermittent generation of the Wind Turbines increased by [2]%.
# Had this increase not occured, the aggregate additional amount of natural gas combusted would have been y Bcf lower, 
# and the corresponding mass of carbon dioxide emitted would have been z million tons lower.
# If this amount of z million tons is allocated to the aggregate wind generation of x GWh, it corresponds 
# to q tons/MWh: this is y% lower than the p tons/MWh for a modern combined-cycle gas fired power plant.

import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None 
import matplotlib.pyplot as plt
selection = [0,1,5,6,8,13,14,94,95]
col_names = ['site','CHP', 'operator', 'state', 'NERC', 'generator', 'fuel', 'Btu', 'MWh']
data_types = {'site': str,'CHP': str, 'operator': str, 'state': str, 'NERC': str , 
             'generator': str, 'fuel': str, 'Btu': int , 'MWh': int}
df10= pd.read_csv("EIA2010.csv", skiprows=8, header= None, names= col_names, usecols= selection,index_col= 'site',
                  dtype = data_types, thousands = ',')# gets rid of the commas denoting thousands: Btu/MWh now type int

df16 = pd.read_csv("EIA2016.csv", skiprows=6, header= None, names= col_names, usecols= selection, index_col= 'site',
                   dtype = data_types, thousands = ',') 
df16.head(4)
df10.head(4)
```

```python
# Convert the remaining variables to Pandas type category
for col in ['CHP', 'operator', 'state', 'NERC', 'fuel','generator']:# more concise w/o repetition
    df10[col] = df10[col].astype('category')
    df16[col] = df16[col].astype('category')
df10.dtypes
df16.dtypes
```

```python
# address missing values (common for CA since often zero): are now imputed as zeros, reduce number for display
df16["Btu"].fillna(0, inplace=True), df10["MWh"].fillna(0, inplace=True)
df16["MWh"].fillna(0, inplace=True), df16["MWh"].fillna(0, inplace=True)
df16.Btu = df16.Btu/1000 
df16.Btu = df16.Btu.astype(int)
df16.MWh = df16.MWh/1000 
df16.MWh = df16.MWh.astype(int)
df16.head(4)
df10["Btu"].fillna(0, inplace=True), df10["MWh"].fillna(0, inplace=True)
df10["MWh"].fillna(0, inplace=True), df16["MWh"].fillna(0, inplace=True)
df10.Btu = df10.Btu/1000 
df10.Btu = df10.Btu.astype(int)
df10.MWh = df10.MWh/1000 
df10.MWh = df10.MWh.astype(int)
df10.head(4)
```

```python
m_CHP = df10.CHP == "N" # m_ abbreviation  for "mask"
m_NERC = df10.NERC == "TRE" # Code for ERCOT: 85% only of TX is in ERCOT
m_fuel= df10.fuel == "NG"
m_CA = df10.generator== 'CA'# a CCGT is an integrated unit with usually one "CT" (Combustion Turbine) and one "CA".
m_CT = df10.generator== 'CT'# we need to seperate out the CT and its matching CA and then re-integrate them as a unit.
CC10 = df10[m_CHP & m_NERC & m_fuel & (m_CA | m_CT)] # CC denotes Combined Cycle (CT & CA) Gas Turbine
CC10.head(8)
len(CC10)
m_CHP = df16.CHP == "N" # We are overwriting the five prior m values and so the composite boolean 5 lines below is OK. 
m_NERC = df16.NERC == "TRE" 
m_fuel= df16.fuel == "NG"
m_CA = df16.generator== 'CA'
m_CT = df16.generator== 'CT'
CC16 = df16[m_CHP & m_NERC & m_fuel & (m_CA | m_CT)] 
CC16.head(8)
#len(CC16)
```

```python
CC10.drop(CC10.columns[[0,2,3,5]], axis=1, inplace=True) # drop now-redundant columns before the CT/CA Merge
CC16.drop(CC16.columns[[0,2,3,5]], axis=1, inplace=True) 
CC10  #.head(4)
```

```python
CA10= CC10[CC10.generator == "CA"]
CT10= CC10[CC10.generator == "CT"]
CT10.info()
# CACT10= CT10.merge(CA10, how = "inner")# , suffixes= ("_CT", "_CA) didnt Run!
CACT10 = pd.merge(CT10, CA10, left_index=True, right_index=True)
CACT10 # Since the match is indeed perfect we can drop now redundant columns and relabel the remaining columns
CACT10.drop(CACT10.columns[[1,4,5]], axis=1, inplace=True)
CACT10
column_indices = [0,1,2,3,4]# This dict approach does work
new_names = ["operator", "Btu_CT","MWh_CT", "Btu_CA","MWh_CA"]
old_names = CACT10.columns[column_indices]
CACT10.rename(columns=dict(zip(old_names, new_names)), inplace=True)  
CACT10
```

```python
CA16= CC16[CC16.generator == "CA"]
CT16= CC16[CC16.generator == "CT"]
CT16.info()
CACT16 = pd.merge(CT16, CA16, left_index=True, right_index=True)
CACT16 # Since the match is indeed perfect we can drop now redundant columns and relabel the remaining columns
CACT16.drop(CACT16.columns[[1,4,5]], axis=1, inplace=True)
CACT16
column_indices = [0,1,2,3,4]# This dict approach does work
new_names = ["operator", "Btu_CT","MWh_CT", "Btu_CA","MWh_CA"]
old_names = CACT16.columns[column_indices]
CACT16.rename(columns=dict(zip(old_names, new_names)), inplace=True)  
CACT16
# CACTdf.head(15) # Later we will repeat for the Year 2010, and then Merge the two DF's for 2010 and 2016.
# Note that plant 12 is an outlier, being very tiny (MWh output is only about 1% of the larger plants, so immaterial)
# DWW: Are TX CCGT plants now flexing w Duct Burners, so Btu_CA/Btu_CT is now higher than 2010?? Map of plants w join?
```
