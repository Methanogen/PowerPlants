```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
selection = [0,1,5,6,8,13,14,94,95]
col_names= ['Site#', 'CHP', 'Operator#', 'State', 'NERC', 'Generator', 'Fuel', 'Btu', 'MWh']
data_types= {'Site#': int, 'CHP': str, 'Operator#': int, 'State': str, 'NERC': str , 
             'Generator': str, 'Fuel': str, 'Btu': str , 'MWh': str}
df= pd.read_csv("EIA2016.csv", skiprows=6, header= None, names= col_names, usecols= selection, dtype= data_types) 
# dropped: low_memory= False, since data types are specified in the dictionary / dropna(how= all) also not used
df.head(3)
```

```python
# convert Btu and MWh to numerical variables after getting rid of the commas
df.dtypes
df.Btu= df.Btu.str.replace( ',', '') #get rid of commas in strings 
df.MWh= df.MWh.str.replace( ',', '') # 
df.head(3)
df.dtypes
df.Btu= df.Btu.astype(int)# convert from string to an integer
df.MWh= df.MWh.astype(int)
df.dtypes
```

```python
df["Btu"].fillna(0, inplace=True)# missing values (common for CA since often zero) are imputed as zeros
df["MWh"].fillna(0, inplace=True)
```

```python
# Convert the remaining variables to Pandas type category
df["Site#"]= df["Site#"].astype('category')
df["CHP"]= df["CHP"].astype('category')
df["Operator#"]= df["Operator#"].astype('category')
df["State"]= df["State"].astype('category')
df["NERC"]= df["NERC"].astype('category')
df["Fuel"]= df["Fuel"].astype('category')
df["Generator"]= df["Generator"].astype('category')
df.dtypes
```

```python
mask1= df.CHP == "N"
mask2= df.State == "TX"
mask3= df.Fuel == "NG"
mask4= df.Generator== 'CA'# a CCGT is an integrated unit with usually One CT (Combustion Turbine) and one CA.
mask5= df.Generator== 'CT'# we need to seperate out the CT and its matching CA and then re-integrate them as a unit.
CCTX= df[mask1 & mask2 & mask3 & (mask4 | mask5)]
CCTX.head(8)
CCTX.drop(CCTX.columns[[1,3,4,6]], axis=1, inplace=True) # drop now-redundant columns before the CT/CA Merge
#CCTX_small= CCTX[:,[CCTX.Site#, CCTX.Operator#]]  #,CCTX[""]"]",2,5,6,7,8]]  #TypeError: unhashable type (??)
CCTX.head(4)  #.head(4)
```

```python
# perhaps first set Site# as the Index?
CAdf= CCTX[CCTX["Generator"]== "CA"]
len(CAdf) # 43 plants w CA
CAdf.head(4)
CTdf= CCTX[CCTX["Generator"]== "CT"]
len(CTdf)  # also 43 plants w CT, and perfect matches on "Site#" seem to be the case (Phew! so Site# is safe as the index!)
CTdf.head(4)
CTdf.info()
CACTdf= CTdf.merge(CAdf, how = "inner", on= "Site#")# , suffixes= ("_CT", "_CA) didnt Run!
CACTdf.head(40)# Since the match is indeed perfect we can drop columns and relabel remaining columns
CACTdf.drop(CACTdf.columns[[2,5,6]], axis=1, inplace=True)
CACTdf.head(4)
# CACTdf= CACTdf.columns["Site#","Operator#", "Btu_CT","MWh_CT", "Btu_CA","MWh_CA"] # Renaming the columns WRONG
column_indices = [0,1,2,3,4,5]# This dict approach does work
new_names = ["Site#","Operator#", "Btu_CT","MWh_CT", "Btu_CA","MWh_CA"]
old_names = CACTdf.columns[column_indices]
CACTdf.rename(columns=dict(zip(old_names, new_names)), inplace=True)                    
```
